# Volatility Model : from implied volatility to local volality using Dupire's equation

This notebook shows how to display the implied volatility surface and computes the local volatiltiy surface, using AssetPricing functions.
The implied volatility is a volatility model stating that unlike the assumptions of Black,Scholes and Merton, the underlying volatility of an asset isn't constant. It rather states that the volatility vary over time and over the spot price. This mean sigma is a function of S and T, often wrote $\sigma$(S,T). 
The local volatility can be computed using different formula, with one being Dupire's formula : 

dσ²(K,T)/dT = 2r(K-rc) dσ²(K,T)/dK² + 2σ(K,T)² K² d²σ²(K,T)/dK²dT

Where:

    σ is the local volatility
    K is the strike price
    T is the time to expiration
    r is the risk-free interest rate
    rc is the cost of carry (which can be zero if there are no dividends)
    dσ²(K,T)/dT is the change in local volatility with respect to time
    dσ²(K,T)/dK² is the second derivative of local volatility with respect to strike price
    d²σ²(K,T)/dK²dT is the mixed derivative of local volatility with respect to both strike price and time to expiration.

We now add our dividends, Spot price and Risk-free rate to our data

```python
option_data['Dividend'] = apple.getDiv()
option_data['Spot'] = apple.getPrice()
option_data['Risk-Free Rate'] = risk_free_rate
```

We will now clean our data before actually computing the implied volatility surface. To clean our data, we do the following : 
- we group by expiry, and compute the spread. Then we compute the mean and std of the spread for each expiry. We will only keep options when the spread if lower or equal than the mean + 4 times the std, to avoid too wide spread
- We remove every options which hasn't been traded in the lat 5 business days, considering it's too illiquid
- Following the same logic, we remove every option which volume or open interest is 0
- Lastly, we only keep a few interesting columns

```python
cleaned_data = apple.clean_Option_Data(option_data)
cleaned_data
```

We will now try to build an implied volatility surface. 

First, we will iterate over the set of expiry. For each expiry, we will compute the implied volatility for each strike present in our cleaned data. We will then use those implied volatility to create an interpolation to create a volatility "skew" or "smile" for each expiry. The interpolation method can be choosed, but scipy.dataggrid linear interpolation seems to be the best.

Once we have our sets of skew, we now want to create a smooth surface. This requires another interpolatation, where we will use RBF function. The build_Impl_Vol_Surface will return 3 arrays, containing respectively the set of strike used at each expiry, the list of expiry and finaly the implied volatility associated for each tuple(strike,expiry)

```python
strikes, expiries, impliedVol = apple.build_Impl_Vol_Surface(cleaned_data, OptionTypes.EUROPEAN_PUT)
```

With those arrays, we can now plot our implied volatility surface. This can be done manually altough i have done a specific function to do so

```python
apple.plot_Vol_Surface(strikes, expiries, impliedVol)
```
